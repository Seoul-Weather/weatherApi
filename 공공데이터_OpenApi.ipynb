```python
import pandas as pd
df = pd.read_csv('50.csv')
df = df.dropna()
df = df[df['장소명'] != '서울대공원']
df
```

```python
import requests
import json
import xmltodict
place = []
url = 'http://openapi.seoul.go.kr:8088/5a63556c6c6a756c36386b5a644163/xml/citydata/1/5/' 
for index,row in df.iterrows():
  place.append(row['장소명'])


places= []
for i in place:
  finalUrl = 'http://openapi.seoul.go.kr:8088/5a63556c6c6a756c36386b5a644163/xml/citydata/1/5/'+i 
  response = requests.get(finalUrl)
  jsonR = json.dumps(xmltodict.parse(response.text), indent = 4)
  data = json.loads(jsonR)
  places.append(data['SeoulRtd.citydata']['CITYDATA']['COVID_19_STTS']['COVID_19_STTS']["GU_NM"])
```

```python
len(places)
```

```python
df['구'] = places
```
